```python
# https://help.naver.com/service/5626/contents/959?lang=ko
query = 'nba' "+연봉"
#query = '한국폴리텍대학 "인공지능"'

url = 'https://www.naver.com/'
driver = webdriver.Chrome()

driver.get(url)
#time.sleep(3)
driver.implicitly_wait(time_to_wait=10)
time.sleep(3)

search_box = driver.find_element(By.CSS_SELECTOR, "input#query")
search_box.send_keys(query)
search_box.send_keys(Keys.RETURN)
driver.implicitly_wait(time_to_wait=10)
time.sleep(3)

#driver.find_element(By.XPATH, '//*[@id="lnb"]/div[1]/div/div[1]/div/div[1]/div[7]/a').click()
#driver.find_element(By.XPATH, '//*[@id="lnb"]/div[1]/div/ul/li[2]/a').click()
#driver.implicitly_wait(time_to_wait=10)

#driver.find_element(By.XPATH, '//a[text()="뉴스"]').click()
driver.find_element(By.LINK_TEXT, '다음').click()
while True:
    btn = driver.find_element(By.LINK_TEXT, '뉴스')
    if (btn.is_displayed):        
        btn.click()
        driver.implicitly_wait(time_to_wait=10)
        time.sleep(3)
        break        
    else:        
        driver.find_element(By.LINK_TEXT, '다음').click()

page_down(100)
```

```python
filename = "네이버뉴스검색정보_nba_연봉.csv"
f = open(filename, "w", encoding="utf-8-sig", newline="")
writer = csv.writer(f)

attributes = ["신문사", "일자", "제목", "주소", "내용"]
writer.writerow(attributes)

for i in range(5): #while True:
    #홈페이지에서 직접 태그 확인
    #XPATH => CSS_SELECTOR로 변환하면 좋당
    news_titles = driver.find_elements(By.CSS_SELECTOR, ".news_tit")
    news_press = driver.find_elements(By.XPATH, '//*[@class="info_group"]/a[1]')
    news_date = driver.find_elements(By.XPATH, '//*[@class="info_group"]/span[1]')
    news_des = driver.find_elements(By.XPATH, '//*[@class="dsc_wrap"]/a[1]')

    for i in range(len(news_press)):
        press = news_press[i].text
        date = news_date[i].text
        #태그 사이에 글자를 가져오는것 text
        title = news_titles[i].text
        #속성중에서 가져오는 함수 get_attribute
        rink = news_press[i].get_attribute("href")
        des = news_des[i].text
        data_rows = [press, date, title, rink, des]
        writer.writerow(data_rows)        
        print(data_rows)

    #driver.find_element(By.CSS_SELECTOR, ".btn_next").click()
    #time.sleep(2)
    driver.implicitly_wait(time_to_wait=1000)

driver.close()
f.close()
```
