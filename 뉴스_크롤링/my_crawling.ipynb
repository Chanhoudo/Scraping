```python
import csv
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import requests
import time

# 특수 문자 제거 함수
def sanitize_filename(filename):
    return re.sub(r'[\\/*?:"<>|.]', "", filename)

def download_image(image_url, image_name, image_folder):
    image_name = sanitize_filename(image_name)  # 파일 이름에서 특수 문자 제거
    image_path = os.path.join(image_folder, image_name)
    if not os.path.exists(image_folder):
        os.makedirs(image_folder)  # 경로가 없으면 생성
    response = requests.get(image_url)
    if response.status_code == 200:
        with open(image_path, 'wb') as f:
            f.write(response.content)
    return image_name

# 네이버에서 딥페이크 뉴스 크롤링
def crawl_naver_news(query='딥페이크', image_folder='static/images', csv_file='딥페이크_검색_결과.csv'):
    url = 'https://www.naver.com/'
    driver = webdriver.Chrome()

    # 네이버 페이지 열기
    driver.get(url)
    driver.implicitly_wait(10)
    time.sleep(3)

    # 검색창에 검색어 입력 후 검색 실행
    search_box = driver.find_element(By.CSS_SELECTOR, "input#query")
    search_box.send_keys(query)
    search_box.send_keys(Keys.RETURN)
    driver.implicitly_wait(10)
    time.sleep(3)

    # 뉴스 탭으로 이동
    while True:
        try:
            btn = driver.find_element(By.LINK_TEXT, '뉴스')
            if btn.is_displayed():
                btn.click()
                driver.implicitly_wait(10)
                time.sleep(3)
                break
        except:
            print("뉴스 버튼을 찾을 수 없음, 다시 시도")
            driver.find_element(By.LINK_TEXT, '다음').click()

    # 스크롤 내리기
    for _ in range(10):
        driver.find_element(By.TAG_NAME, "body").send_keys(Keys.PAGE_DOWN)
        time.sleep(0.2)

    # CSV 파일 준비
    f = open(csv_file, "w", encoding="utf-8-sig", newline="")
    writer = csv.writer(f)
    attributes = ["신문사", "일자", "제목", "주소", "내용", "이미지 경로"]
    writer.writerow(attributes)

    # 뉴스 데이터 수집
    news_titles = driver.find_elements(By.CSS_SELECTOR, ".news_tit")
    news_press = driver.find_elements(By.XPATH, '//*[@class="info_group"]/a[1]')
    news_date = driver.find_elements(By.XPATH, '//*[@class="info_group"]/span[1]')
    news_des = driver.find_elements(By.XPATH, '//*[@class="dsc_wrap"]/a[1]')
    images = driver.find_elements(By.CSS_SELECTOR, ".dsc_thumb img")

    for i in range(len(news_press)):
        press = news_press[i].text
        date = news_date[i].text
        title = news_titles[i].text
        link = news_titles[i].get_attribute("href")
        des = news_des[i].text

        # 이미지 URL 추출 및 다운로드
        image_url = images[i].get_attribute("src") if i < len(images) else None
        if image_url:
            image_name = title.replace(" ", "_") + '.jpg'
            image_name = download_image(image_url, image_name, image_folder)
        else:
            image_name = "이미지 없음"

        # 수집한 데이터를 CSV에 저장
        data_rows = [press, date, title, link, des, image_name]
        writer.writerow(data_rows)

    driver.close()
    f.close()
    print("크롤링 완료")
```

```python
from flask import Flask, render_template
import csv
import os

app = Flask(__name__)

# CSV 파일을 읽고 데이터를 리스트로 반환하는 함수
def read_csv():
    news_data = []
    csv_file = '딥페이크_검색_결과.csv'
    
    with open(csv_file, 'r', encoding='utf-8-sig') as file:
        reader = csv.reader(file)
        next(reader)  # 첫 번째 줄은 헤더이므로 건너뜀
        for row in reader:
            news_data.append({
                'press': row[0],
                'date': row[1],
                'title': row[2],
                'link': row[3],
                'description': row[4],
                'image_path': row[5]
            })
    return news_data

# 메인 페이지 라우팅
@app.route('/')
def index():
    news_data = read_csv()
    return render_template('index.html', news_data=news_data)
# 로컬에 저장된 이미지를 제공하는 라우트
@app.route('/image/<filename>')
def image(filename):
    return send_from_directory(app.config['IMAGE_FOLDER'], filename)

if __name__ == '__main__':
    app.run(debug=True, use_reloader=False)
```
